```python
import requests
from PIL import Image
import io
from IPython.display import display, Markdown
```

```python
# Base URL for the Flask app
base_url = "http://127.0.0.1:5000"
```

```python
# API Instructions from the '/' endpoint
# Send GET request to the '/' endpoint
response = requests.get(base_url + "/")
display(Markdown("## API Instructions"))
display(Markdown("### Response JSON:"))
display(response.json())
```

```python
# Send POST request to the `/convert` endpoint with image
with open("isagi.jpg", "rb") as image_file:
    files = {"file": image_file}  # 'file' matches the expected parameter name
    data = {"format": "PNG"}  # Desired output format, adjust as needed

    # Send the request
    response = requests.post(base_url + "/convert", files=files, data=data)

    # Display the converted image if the request was successful
    if response.status_code == 200:
        img = Image.open(io.BytesIO(response.content))
        display(Markdown("## Converted Image"))
        display(img)
    else:
        display(Markdown(f"**Error:** {response.json()}"))
```

```python
# Demonstrate the `/fake_user_agent` Endpoint
# Send GET request to the `/fake_user_agent` endpoint with browser parameter
params = {"browser": "chrome"}
response = requests.get(base_url + "/fake_user_agent", params=params)
```

```python
display(Markdown("## Fake User-Agent"))
if response.status_code == 200:
    display(Markdown(f"**Generated User-Agent:** {response.json()['user_agent']}"))
else:
    display(Markdown(f"**Error:** {response.json()}"))
```
